```python
URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
spacex_df = pd.read_csv(URL)
spacex_df
```

```python
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class', 'Landing Outcome']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]
launch_sites_df
```

```python
# Start location is NASA Johnson Space Center
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=10)
site_map
```

```python
# Creates a blue circle at NASA Johnson Space Center's coordinate with a popup label showing its name
circle = folium.Circle(nasa_coordinate, radius=1000, color='#d35400', fill=True).add_child(folium.Popup('NASA Johnson Space Center'))
# Creates a blue circle at NASA Johnson Space Center's coordinate with a icon showing its name
marker = folium.map.Marker(
    nasa_coordinate,
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'NASA JSC',
        )
    )
site_map.add_child(circle)
site_map.add_child(marker)
```

```python
import folium
from folium import DivIcon

# Initialize the map
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Iterates through the launch_sites DataFrame
for index, row in launch_sites_df.iterrows():
    # Get the coordinates for the launch site
    coordinate = [row['Lat'], row['Long']]

    # Create a folium.Marker for the launch site
    marker = folium.map.Marker(
        coordinate,
        # Create an icon as a text label
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % row['Launch Site'],
            )
        )

    # Create a folium.Circle for the launch site
    circle = folium.Circle(coordinate, radius=1000, color='#d35400', fill=True).add_child(folium.Popup(row['Launch Site']))

    # Add the folium.Marker and folium.Circle to the map
    site_map.add_child(marker)
    site_map.add_child(circle)


# Display the map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
distance_coastline = calculate_distance(28.573255, -80.646895, 28.62504, -80.70852)
distance_coastline
```

```python
# Defines the launch site coordinates and the coastline coordinates
launch_site_lat = 28.573255
launch_site_lon = -80.646895
coastline_lat = 28.62504
coastline_lon = -80.70852

# Calculates the distance between the launch site and the coastline
distance = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
coordinates = [(launch_site_lat, launch_site_lon), (coastline_lat, coastline_lon)]

# Creates a folium.Marker on the coastline point on the map
distance_marker = folium.Marker(
    [coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
        )
    )

# Adds the distance marker to the map
site_map.add_child(distance_marker)
lines = folium.PolyLine(locations=coordinates, weight=1)
site_map.add_child(lines)

# Display the map
site_map
```
